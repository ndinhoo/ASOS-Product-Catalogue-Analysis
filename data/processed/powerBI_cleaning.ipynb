```python
import pandas as pd 
import regex as re
```

```python
df = pd.read_csv("../powerBI/powerbi_products_asos.csv")
```

```python
df = df.dropna()

#getting base colors cause many colors
def base_color(color):
    if "green" in color:
        return "green"
    elif "blue" in color:
        return "blue"
    elif "grey" in color:
        return "grey"
    elif "white" in color:
        return "white"
    elif "grey" in color:
        return "grey"
    elif "pink" in color:
        return "pink"
    elif "purple" in color:
        return "purple"
    elif "orange" in color:
        return "orange"
    elif "brown" in color:
        return "brown"
    elif "red" in color:
        return "red"
    elif "khaki" or "beige" in color:
        return "beige"
    
    
df["base_color"] = df["color"].apply(base_color)
    
#clothes without stock
df["has_out_of_stock"] = df["size"].str.contains("Out of stock")

nb_total       = []
nb_available   = []
nb_out         = []
has_out        = []
fully_out      = []

for size_str in df['size']:
    parts = size_str.split(',')

    available    = 0
    out_of_stock = 0

    for part in parts:
        if 'out of stock' in part.lower():
            out_of_stock += 1
        else:
            available += 1

    nb_total.append(len(parts))
    nb_available.append(available)
    nb_out.append(out_of_stock)
    has_out.append(out_of_stock > 0)
    fully_out.append(available == 0)

df['nb_sizes_total']        = nb_total
df['nb_sizes_available']    = nb_available
df['nb_sizes_out_of_stock'] = nb_out
df['has_out_of_stock']      = has_out
df['fully_out_of_stock']    = fully_out

#price segment 
price_segment = []

for price in df['price']:
    if price < 20:
        price_segment.append('<£20')
    elif price < 50:
        price_segment.append('£20-50')
    elif price < 100:
        price_segment.append('£50-100')
    else:
        price_segment.append('>£100')

df['price_segment'] = price_segment

df = df.drop(columns=['size'])
df.to_csv('/Users/fefe/Desktop/Trendsetter/notebook/powerBI/asos_clean.csv', index=False)
```
